```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tools.eval_measures import mse, rmse
from pmdarima import auto_arima
%matplotlib inline
```

```python
co2_mm=pd.read_csv('../Data/co2_mm_mlo.csv') 
print(co2_mm.shape,co2_mm.columns)
co2_mm.head()
```

```python
co2_mm['date']=pd.to_datetime({'year':co2_mm['year'],'month':co2_mm['month'], 'day':1})
co2_mm.info()
co2_mm.head()
```

```python
co2_mm=co2_mm.set_index('date')
co2_mm.index.freq='MS'
co2_mm.head()
```

```python
co2_mm['interpolated'].plot(figsize=(12,5));
```

```python
co2_mm_decompose=seasonal_decompose(co2_mm['interpolated'],model='add')
co2_mm_decompose.plot();
```

```python
co2_mm_decompose.seasonal.plot(figsize=(12,5));
```

```python
co2_mm_aa_model=auto_arima(co2_mm['interpolated'], seasonal=True, m=12, trace=True, error_action='ignore', suppress_warnings=True)
co2_mm_aa_model.summary()
```

```python
co2_mm_aa_model.summary()
```

```python
train=co2_mm[:717]
test=co2_mm[717:]
print(train.shape,test.shape)
```

```python
model = SARIMAX(train['interpolated'], order=(0, 1, 3), seasonal_order=(1, 0, 1, 12))
result = model.fit()
result.summary()
```

```python
start_=len(train)
end_=start_+len(test)-1
pred=result.predict(start=start_, end=end_, typ='levels').rename('SARIMA Ord (0, 1, 3) Seasonal Ord(1, 0, 1, 12)')
```

```python
test['interpolated'].plot(figsize=(12,5), legend= True)
pred.plot(legend= True);
```

```python
error_rmse=rmse(test['interpolated'],pred)
error_rmse
```

```python
test['interpolated'].mean()
```

```python
pred.mean()
```

```python
model = SARIMAX(co2_mm['interpolated'], order=(0, 1, 3), seasonal_order=(1, 0, 1, 12))
result = model.fit()
print(result.summary())
fcast=result.predict(start=len(co2_mm), end=len(co2_mm)+11, typ='levels').rename('SARIMA Forecast')
```

```python
co2_mm['interpolated'].plot(figsize=(12,5),legend=True)
fcast.plot(legend=True);
```
