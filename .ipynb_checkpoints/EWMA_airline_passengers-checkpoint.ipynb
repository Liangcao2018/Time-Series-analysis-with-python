```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
airline=pd.read_csv('../Data/airline_passengers.csv',index_col='Month', parse_dates=True)
airline.head()
```

```python
airline.info()
```

```python
airline.index
```

```python
airline['6_month_SMA']=airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12_month_SMA']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.plot(figsize=(12,5));
```

```python
airline['EWMA_12']=airline['Thousands of Passengers'].ewm(span=12).mean()
airline[['Thousands of Passengers','EWMA_12']].plot(figsize=(12,8));
```
