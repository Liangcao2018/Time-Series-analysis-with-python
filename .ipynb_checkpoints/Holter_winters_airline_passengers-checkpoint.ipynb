```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
%matplotlib inline
```

```python
airline=pd.read_csv('../Data/airline_passengers.csv',index_col='Month', parse_dates=True)
airline.head()
```

```python
airline.info()
```

```python
airline.index
```

```python
airline.index.freq='MS'
```

```python
airline.index
```

```python
airline.head()
```

```python
span_ = 12
alpha_ = 2/(span_+1)
```

```python
airline['EWMA_12']=airline['Thousands of Passengers'].ewm(alpha=alpha_,adjust=False).mean()
airline.head(6)
```

```python
model = SimpleExpSmoothing(airline['Thousands of Passengers'])
model_trained = model.fit(smoothing_level=alpha_,optimized=False)
```

```python
model_trained.fittedvalues.head(6)
```

```python
model_trained.fittedvalues.shift(-1).head(6)
```

```python
airline['SES12']=model_trained.fittedvalues.shift(-1)
airline.head()
```

```python
airline['DB_add_12']=ExponentialSmoothing(airline['Thousands of Passengers'],trend='add').fit().fittedvalues.shift(-1)
airline['DB_mul_12']=ExponentialSmoothing(airline['Thousands of Passengers'],trend='mul').fit().fittedvalues.shift(-1)
airline.head(6)
```

```python
airline.columns
```

```python
airline[['Thousands of Passengers', 'SES12', 'DB_add_12', 'DB_mul_12']].plot(figsize=(12,8));
```

```python
%%time
airline['TES_mul_12']=ExponentialSmoothing(airline['Thousands of Passengers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues.shift(-1)
```
