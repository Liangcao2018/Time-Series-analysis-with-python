```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.api import VAR
from statsmodels.tsa.statespace.varmax import VARMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse

%matplotlib inline
```

```python
def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
M2S=pd.read_csv('../Data/M2SLMoneyStock.csv',index_col=0,parse_dates=True) 
print(M2S.shape,M2S.columns)
M2S.index.freq='MS'
M2S.head()
```

```python
PS=pd.read_csv('/opsmnt/Projects/TSA_COURSE_NOTEBOOKS/Data/PCEPersonalSpending.csv',index_col=0,parse_dates=True) 
print(PS.shape,PS.columns)
PS.index.freq='MS'
PS.head()
```

```python
DF_comb=pd.concat([PS,M2S],axis=1)
print(DF_comb.shape,DF_comb.columns)
DF_comb.index
```

```python
DF_comb=DF_comb.dropna()
print(DF_comb.shape,DF_comb.columns)
DF_comb.head()
```

```python
DF_comb.plot(figsize=(12,5));
```

```python
adf_test(DF_comb['Spending'])
```

```python
DF_comb_tranform=DF_comb.diff()
```

```python
adf_test(DF_comb_tranform['Spending'])
```

```python
adf_test(DF_comb_tranform['Money'])
```

```python
DF_comb_tranform=DF_comb_tranform.diff()
DF_comb_tranform=DF_comb_tranform.dropna()
DF_comb_tranform.shape
```

```python
adf_test(DF_comb_tranform['Spending'])
```

```python
adf_test(DF_comb_tranform['Money'])
```

```python
DF_comb_tranform.plot(figsize=(12,5));
```

```python
nobs=12
train=DF_comb_tranform[:-nobs]
test=DF_comb_tranform[-nobs:]
print(train.shape,test.shape)
train.tail()
```

```python
test.head()
```

```python
model = VAR(train)
result_aic=[]
for p in range(1,8):
    result = model.fit(p)    
    print(f'Order : {p} , AIC : {result.aic}')
    result_aic.append(result.aic)

result_aic=np.array(result_aic)
print(f'Lowest AIC : {np.min(result_aic)} for p value: {np.argmin(result_aic)+1}')
```

```python
result = model.fit(np.argmin(result_aic)+1)
result.summary()
```

```python
lagged_values = train.values[-5:]
```

```python
test.index
```

```python
z = result.forecast(y=lagged_values, steps=12)
idx=pd.date_range(start='2015-01-01', periods=12, freq='MS')
var_forecast=pd.DataFrame(z, index=idx, columns=['Spending_2d','Money_2d'])
var_forecast
```

```python
var_forecast['Money_1d'] = (DF_comb['Money'].iloc[-nobs-1] - DF_comb['Money'].iloc[-nobs-2]) + var_forecast['Money_2d'].cumsum()
var_forecast['Money_forecast'] = DF_comb['Money'].iloc[-nobs-1] + var_forecast['Money_1d'].cumsum()
```

```python
var_forecast['Spending_1d'] = (DF_comb['Spending'].iloc[-nobs-1] - DF_comb['Spending'].iloc[-nobs-2]) + var_forecast['Spending_2d'].cumsum()
var_forecast['Spending_forecast'] = DF_comb['Spending'].iloc[-nobs-1] + var_forecast['Spending_1d'].cumsum()
```

```python
var_forecast
```

```python
DF_comb[-nobs:].plot(figsize=(12,8), legend= True);
```

```python
var_forecast[['Spending_forecast','Money_forecast']].plot(figsize=(12,8),legend= True);
```

```python
rmse_error_Money=rmse(DF_comb[-nobs:]['Money'],var_forecast['Money_forecast'])
print(f'RMSE : {rmse_error_Money}')
Money_act_avg=DF_comb[-nobs:]['Money'].mean()
Money_fct_avg=var_forecast['Money_forecast'].mean()
print(f'Money actual mean:{Money_act_avg},Money forecast mean:{Money_fct_avg}')
DF_comb[-nobs:]['Money'].plot(figsize=(12,8), legend= True)
var_forecast['Money_forecast'].plot(figsize=(12,8),legend= True);
```

```python
rmse_error_spending=rmse(DF_comb[-nobs:]['Spending'],var_forecast['Spending_forecast'])
print(f'RMSE : {rmse_error_spending}')
Spending_act_avg=DF_comb[-nobs:]['Spending'].mean()
Spending_fct_avg=var_forecast['Spending_forecast'].mean()
print(f'Spending actual mean:{Spending_act_avg},Spending forecast mean:{Spending_fct_avg}')
DF_comb[-nobs:]['Spending'].plot(figsize=(12,8), legend= True)
var_forecast['Spending_forecast'].plot(figsize=(12,8),legend= True);
```
