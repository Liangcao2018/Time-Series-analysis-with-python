```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
BWL_df=pd.read_csv('../Data/BeerWineLiquor.csv')
BWL_df.head()
```

```python
BWL_df.info()
```

```python
BWL_df.columns=['ds','y']
BWL_df.head()
```

```python
BWL_df['ds'] = pd.to_datetime(BWL_df['ds'])
BWL_df.info()
```

```python
BWL_df.head()
```

```python
m = Prophet()
m.fit(BWL_df)
```

```python
future = m.make_future_dataframe(periods=24, freq='MS')
```

```python
future.head(10)
```

```python
future.tail(5)
```

```python
len(BWL_df)
```

```python
len(future)
```

```python
forecast = m.predict(future)
```

```python
forecast.tail()
```

```python
forecast.columns
```

```python
forecast[['ds', 'yhat_lower', 'yhat', 'yhat_upper']]
```

```python
m.plot(forecast);
```
